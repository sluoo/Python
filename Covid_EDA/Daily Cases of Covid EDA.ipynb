```python
import numpy as np 
import pandas as pd
import os 
import urllib
```

```python
url = " https://covid19.who.int/WHO-COVID-19-global-data.csv"
file_path = os.path.join(os.getcwd(),"data","covid") #create path
os.makedirs(file_path,exist_ok=True) #make directories 

csv_path = os.path.join(file_path,"WHO-COVID-19-global-data.csv")
urllib.request.urlretrieve(url,csv_path) #put url into this path
df = pd.read_csv(csv_path) #read path into df object 
```
