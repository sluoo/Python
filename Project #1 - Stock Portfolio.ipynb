```python
#Import the libraries 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import random 
import os 
```

```python
#Directory 
os.getcwd()
```

```python
#Load csv data 
df = pd.read_excel('/Users/sherry/Desktop/myportfolio.xlsx')
df
```

```python
#Get the total invested amount of money in the portfolio
portfolio_totalAmount = round(sum(df['Amount']*df['Average_Price']),2)
portfolio_totalAmount 

```

```python
#Visualize the porfolio and some additional information
stockTickers = df['Ticker'].values 
size = df['Amount'] * df['Average_Price'] #contribution amount

listZeros = [0]*df.shape[0] #df.shape[0], dim of df 
n = random.randint(0,df.shape[0]-1)
listZeros[n] = 0.1 # explode one slice for pichart 
explode = listZeros
```

```python
print(listZeros)
print(n)
```

```python
#Create figure 
fig1 , ax1 = plt.subplots(figsize= (10,10))
#Plot
ax1.pie(size,explode=explode,labels=stockTickers,shadow='True',autopct='%.2f%%',startangle=360)
#Set title
ax1.set_title('Portfolio Pie Chart',color='Purple',fontsize=22)
#Add text to visual
x = -1.75
y = 1
ax1.text(x,y,"Overview:",fontsize=24,color='Purple')
ax1.text(x,y-0.15,"Total:"+str(portfolio_totalAmount),fontsize=16,color='Blue')

#Place amount under overview 
y_counter = 0.30
for i in range (0,df.shape[0]): 
    ax1.text(x, y - y_counter ,df['Ticker'][i] +": $"+ str(round(df['Amount'][i]*df['Average_Price'][i],2)),
             fontsize=14, color='Black')
    y_counter = y_counter + 0.12 



```
